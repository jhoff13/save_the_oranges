```python
import py2Dmol
import pandas as pd
```

```python
sheet_id = "1nj_Q4yW3CKamMm7nLZ8RuO-9RDXBVSBbQouMi3Asohs"
gid = 1207383717

url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv&gid={gid}"

targets_df = pd.read_csv(url)
targets_df.dropna(subset=['Gene Name:'],inplace=True)
targets_df.head()
```

```python
# All Targets:
target_array = targets_df['AlphafoldDB Link:'].values
with py2Dmol.grid(4, round(target_array.shape[0]/4), size=(300,300), box=True) as g:
  for target in target_array:
      g.view(rotate=True).from_afdb(target, color="plddt")
```

```python
#By Class:
for Class in targets_df['Target Category:'].unique():
  temp_df = targets_df[targets_df['Target Category:'] == Class]
  print('-'*100,f'\n>{Class}')
  target_array = temp_df['AlphafoldDB Link:'].values
  with py2Dmol.grid(4, round(target_array.shape[0]/4), size=(300,300), box=True) as g:
    for target in target_array:
        g.view(rotate=True).from_afdb(target, color="plddt")
```
